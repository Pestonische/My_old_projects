# Распознавание образов

Разработаем метрический алгоритм распознавания, использующий в себе различные подходы для выбора функции схожести среди предложенных вариантов метрической близости, для выбора функции схожести объекта с объектами конкретного класса и для решающего правила

## Постановка задачи

Попытаемся классифицировать картинки из датасета CIFAR10 - представляет собой набор изображений, которые обычно используются для обучения алгоритмов машинного обучения и компьютерного зрения . Это один из наиболее широко используемых наборов данных для исследований в области машинного обучения. Набор данных CIFAR-10 содержит 60 000 цветных изображений 32x32 в 10 различных классах. 10 различных классов представляют самолеты (`plane`), автомобили (`car`), птиц (`bird`), кошек (`cat`), оленей (`deer`), собак (`dog`), лягушек (`frog`), лошадей (`horse`), корабли (`ship`) и грузовики (`track`). Есть 6000 изображений каждого класса.

Мы возьмем некоторую подвыборку из этого датасета. По ней и попытаемся реализовать обучение и тестирование нашего алгоритма. Сразу заметим, что с помощью метрического алгоритма мы, увы, не сможем добиться существенной точности предсказаний

В качестве функции попарной схожести будем использовать следующие варианты:
1. Метрика Евклида    $s(x_1, x_2) = \sqrt{\sum_{i=1}^{n}(x_{1i} - x_{2i})^2}$
2. Метрика Хемминга   $s(x_1, x_2) = \sum_{i=1}^{n}|x_{1i} - x_{2i}|$
3. Метрика Минковского   $s(x_1, x_2) = (\sum_{i=1}^{n}{(x_{1i} - x_{2i})^p})^{\frac{1}{p}}, p \in \mathbb{N}$

В качестве функции схожести объекта выборки с классом будем использовать следующие варианты:
1. Среднее расстояние до класса $X_i$   $f_i(x) = \frac{1}{m_i}\sum_{x_j \in X_{i(o)}^0}s(x, x_j)$
2. По минимальному расстоянию до объектов класса $X_i$   $f_i(x) = \min_{x_j \in X_{i(o)}^0} s(x, x_j)$

В качестве решающего правила возьмем:
1. По минимуму оценки до класса $X_i$   $P_i^A(x) = \begin{cases}
                                                        1, \text{если } f_i(x) = min_{i \in \{1,..,l\}}f_i(x)\\
                                                        0, иначе
                                                    \end{cases} $
                                                     

## Класс для представления метрического алгоритма

```python
import numpy as np
import scipy
from scipy.spatial.distance import cdist
from scipy.special import softmax
import cv2
from sklearn.metrics import accuracy_score
```

```python

class CustomClassifier:
    _estimator_type = "classifier"
    
    def __init__(self, metric="euclidean", p=2, centroids="average", eps=1e-9):
        """K-Nearest Neighbors classifier.
        
        Args:
            n_neighbors: int, default=5
                Number of neighbors to use by default for :meth:`kneighbors` queries.
            metric : {'euclidean', 'minkowski', 'hamming'}
            centroids : {'average', 'min'}
            eps : float, default=1e-5
                Epsilon to prevent division by 0 
        """
        self.metric = metric
        self.centroids = centroids
        self.p = p
        self.eps = eps
    
    def get_pairwise_distances(self, X, Y):
        """
        Returns matrix of the pairwise distances between the rows from both X and Y.
        Args:
            X: numpy array of shape (n_samples, n_features)
            Y: numpy array of shape (k_samples, n_features)
        Returns:
            P: numpy array of shape (n_samples, k_samples)
                Matrix in which (i, j) value is the distance 
                between i'th row from the X and j'th row from the Y.
        """
        return cdist(X, Y, metric=self.metric, p=self.p)
    
    def _get_classes_distances(self, distances_by_classes):
        """
        Returns array of the distances between objects from fit functionand classes in general
        Args:
            distances_by_classes: list of numpy arraya with l elements of shape (k, m_i)
        Returns:
            P: numpy array with shapes (l, k)
        """
        if self.centroids == 'average':
            return np.stack([distances.mean(axis=1) for distances in distances_by_classes])
        elif self.centroids == 'min':
            return np.stack([distances.min(axis=1) for distances in distances_by_classes])
    
    def _get_classes_labels(self, classes_distances):
        return np.argmin(classes_distances, axis=0)
    
    def _get_classes_probs(self, classes_distances):
        return softmax(1 / (classes_distances + eps), axis=1)
            
        
    def fit(self, X, y, classes=None):
        """Fit the model.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
            y: numpy array of shape (n_samples,)
                Target vector.        
        """
        self.points = X
        self.y = y
        #print(X.shape)
        #print((y==1).shape)
        self.points_by_classes = [X[y==label] for label in np.unique(y)]
        if classes:
            self.classes_ = classes
        else:
            self.classes_ = np.unique(y)
        
    
    def predict_proba(self, X):
        """Predict positive class probabilities.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
        Returns:
            y: numpy array of shape (n_samples, n_classes)
                Vector containing positive class probabilities.
        """
        if hasattr(self, 'points'):
            #1. Определение попарных расстояний между объектами тестовой выборки и
            # поклассово объектами тренировачной выборки
            distances_by_classes = [self.get_pairwise_distances(X, train_points) 
                                    for train_points in self.points_by_classes]
            
            #2. Определение схожести объектов с классами в целом
            classes_distances = self._get_classes_distances(distances_by_classes)
            
            #3. Определение вероятности класса объекта
            #labels = self._get_classes_labels(classes_distances)
            probabilities = self._get_classes_probs(classes_distances)
        
        else: 
            raise NotFittedError("CustomClassifier instance is not fitted yet")
            
        
    def predict(self, X):
        """Predict classes.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
        Returns:
            y: numpy array of shape (n_samples,)
                Vector containing predicted class labels.
        """
        #return self.classes_[self.predict_proba(X).argmax(axis=1)]
        if hasattr(self, 'points'):
            #1. Определение попарных расстояний между объектами тестовой выборки и
            # поклассово объектами тренировачной выборки
            distances_by_classes = [self.get_pairwise_distances(X, train_points) 
                                    for train_points in self.points_by_classes]
            #print(len(distances_by_classes), distances_by_classes[0].shape)
            
            #2. Определение схожести объектов с классами в целом
            classes_distances = self._get_classes_distances(distances_by_classes)
            #print(classes_distances.shape)
            
            #3. Определение метки объекта (метка ближайшего класса)
            return self.classes_[self._get_classes_labels(classes_distances)]
        else: 
            raise NotFittedError("CustomClassifier instance is not fitted yet")
```

## Реализация алгоритма, основанного на метрической прецедентности c бинаризацией признаков

```python
class CustomClassifier:
    _estimator_type = "classifier"
    
    def __init__(self, metric="euclidean", p=2, centroids="average", eps=1e-9):
        """K-Nearest Neighbors classifier.
        
        Args:
            n_neighbors: int, default=5
                Number of neighbors to use by default for :meth:`kneighbors` queries.
            metric : {'euclidean', 'minkowski', 'hamming'}
            centroids : {'average', 'min'}
            eps : float, default=1e-5
                Epsilon to prevent division by 0 
        """
        self.metric = metric
        self.centroids = centroids
        self.p = p
        self.eps = eps
    
    def get_pairwise_distances(self, X, Y):
        """
        Returns matrix of the pairwise distances between the rows from both X and Y.
        Args:
            X: numpy array of shape (n_samples, n_features)
            Y: numpy array of shape (k_samples, n_features)
        Returns:
            P: numpy array of shape (n_samples, k_samples)
                Matrix in which (i, j) value is the distance 
                between i'th row from the X and j'th row from the Y.
        """
        return cdist(X, Y, metric=self.metric, p=self.p)
    
    def _get_classes_distances(self, distances_by_classes):
        """
        Returns array of the distances between objects from fit functionand classes in general
        Args:
            distances_by_classes: list of numpy arraya with l elements of shape (k, m_i)
        Returns:
            P: numpy array with shapes (l, k)
        """
        if self.centroids == 'average':
            return np.stack([distances.mean(axis=1) for distances in distances_by_classes])
        elif self.centroids == 'min':
            return np.stack([distances.min(axis=1) for distances in distances_by_classes])
    
    def _get_classes_labels(self, classes_distances):
        return np.argmin(classes_distances, axis=0)
    
    def _get_classes_probs(self, classes_distances):
        return softmax(1 / (classes_distances + eps), axis=1)
            
        
    def fit(self, X, y, classes=None):
        """Fit the model.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
            y: numpy array of shape (n_samples,)
                Target vector.        
        """
        self.points = X
        self.y = y
        #print(X.shape)
        #print((y==1).shape)
        self.points_by_classes = [X[y==label] for label in np.unique(y)]
        if classes:
            self.classes_ = classes
        else:
            self.classes_ = np.unique(y)
        
    
    def predict_proba(self, X):
        """Predict positive class probabilities.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
        Returns:
            y: numpy array of shape (n_samples, n_classes)
                Vector containing positive class probabilities.
        """
        if hasattr(self, 'points'):
            #1. Определение попарных расстояний между объектами тестовой выборки и
            # поклассово объектами тренировачной выборки
            distances_by_classes = [self.get_pairwise_distances(X, train_points) 
                                    for train_points in self.points_by_classes]
            
            #2. Определение схожести объектов с классами в целом
            classes_distances = self._get_classes_distances(distances_by_classes)
            
            #3. Определение вероятности класса объекта
            #labels = self._get_classes_labels(classes_distances)
            probabilities = self._get_classes_probs(classes_distances)
        
        else: 
            raise NotFittedError("CustomClassifier instance is not fitted yet")
            
        
    def predict(self, X):
        """Predict classes.
        
        Args:
            X: numpy array of shape (n_samples, n_features)
        Returns:
            y: numpy array of shape (n_samples,)
                Vector containing predicted class labels.
        """
        #return self.classes_[self.predict_proba(X).argmax(axis=1)]
        if hasattr(self, 'points'):
            #1. Определение попарных расстояний между объектами тестовой выборки и
            # поклассово объектами тренировачной выборки
            distances_by_classes = [self.get_pairwise_distances(X, train_points) 
                                    for train_points in self.points_by_classes]
            #print(len(distances_by_classes), distances_by_classes[0].shape)
            
            #2. Определение схожести объектов с классами в целом
            classes_distances = self._get_classes_distances(distances_by_classes)
            #print(classes_distances.shape)
            
            #3. Определение метки объекта (метка ближайшего класса)
            return self.classes_[self._get_classes_labels(classes_distances)]
        else: 
            raise NotFittedError("CustomClassifier instance is not fitted yet")
```

```python
X_train_a_diff_la = np.append(X_train_a_diff[0].reshape(-1, X_train_a_diff.shape[1]), X_train_a_diff, axis=0)
X_train_a_diff_ra = np.append(X_train_a_diff, X_train_a_diff[-1].reshape(-1, X_train_a_diff.shape[1]), axis=0)
```

```python
X_train_a_diff_la.shape, X_train_a_diff_ra.shape
```

```python
X_train_left = X_train_a - X_train_a_diff_la / 2
X_train_right = X_train_a + X_train_a_diff_ra / 2
```

```python
X_train_intervals = np.zeros((*X_train_a.shape, 2))
X_train_intervals.shape
```

```python
for i in range(X_train_a.shape[0]):
  for j in range(X_train_a.shape[1]):
    for k in range(X_train_a.shape[0]):
      if X_train_a[i, j] >= X_train_left[k, j] and X_train_a[i, j] < X_train_right[k, j]:
        X_train_intervals[i, j] = np.array([k, j])
        break
```

```python
for i in range(X_test_a.shape[0]):
  for j in range(X_test_a.shape[1]):
    for k in range(X_test_a.shape[0]):
      if X_test_a[i, j] >= X_train_left[k, j] and X_test_a[i, j] < X_train_right[k, j]:
        X_test_a_intervals[i, j] = np.array([k, j])
        break
```

```python
X_test_a_intervals.shape
```

```python
def precedent(X_train, X_test):
  precedents = np.zeros((X_test.shape[0], X_train.shape[0]))
  for i in range(X_test.shape[0]):
    for j in range(X_train.shape[0]):
      value = 0
      for k in range(X_test.shape[1]):
        if (X_test[i, k, :] == X_train[j, k, :]).all():
          value += 1
        else:
          value -= 1 
      value /= X_test.shape[1]
      precedents[i, j] = value
  return precedents
```
